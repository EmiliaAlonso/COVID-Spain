# Análisis de Serie de tiempo - Pronostico

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from pylab import rcParams
```

```python
df_serie = df_final.groupby(["fecha"])[["num_casos"]].sum()
df_serie = pd.DataFrame(df_serie)
df_serie.reset_index(inplace=True)
df_serie
```

```python
fig,ax1 = plt.subplots(figsize = (10,3))
serie = df_serie["num_casos"]
serie.index = df_serie["fecha"]
labels = serie.index
ax1.set_title("CASOS POSITIVOS COVID")
ax1.set_xlabel("Fechas")
ax1.set_ylabel("Cantidad personas")
ax1.set_xticklabels(labels)
serie.plot()
plt.show()
```

```python
rcParams["figure.figsize"] = 18,8
plt.figure(figsize = (500,100))
decomposition = sm.tsa.seasonal_decompose(serie,model = "additive")

figura = decomposition.plot()
plt.show()
```

##  Stationarity Analysis

```python
# ADF TEST

from statsmodels.tsa.stattools import adfuller
resultado = adfuller(serie,maxlag = 12, regression = "nc")
print('Estadistico ', resultado[0])
print('p-valor ', resultado[1])
print('Valores Criticos')
for k,v in resultado[4].items():
    print(k,v)
```

La serie es estacionaria

```python
#lets transform our target into more likeable numbers

serie_log = np.log(serie)
```

```python
#estimation
mod = sm.tsa.statespace.SARIMAX(serie,
                                order = (6,2,5),
                                trend=('t'), #trend
                            seasonal_order=(0, 0, 0, 0),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit()
```

```python
print(results.summary())

results.plot_diagnostics(figsize=(16,8))
plt.show()
```

```python
# verification

residuos = results.resid
fig,ax=plt.subplots(2,1,figsize=(15,8))
fig = sm.graphics.tsa.plot_acf(residuos, lags=10, ax=ax[0])
fig = sm.graphics.tsa.plot_pacf(residuos, lags=10, ax=ax[1])
plt.show()
```

Since there is no evidence of remanent correlation between the residuals and the Prob(Q) from Lyung Box (0.99) is greater than the p-value (0.05) we can say this result is stationary, the residuals are white noise.

## Forecast

```python
pred = results.get_forecast(10) 
pred.predicted_mean
```

```python
#interval of predicted values
print(pred.summary_frame())
```

```python
casos_predichos = results.get_prediction().predicted_mean
```

```python
plt.plot(serie , color ="Blue") #real value
plt.plot(casos_predichos, color ="Green") #forecast
#plt.savefig("Resultado PBI vs real.png",format  ="png", dpi = 100, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

# Plot the data (here we are subsetting it to get a better look at the forecasts
serie.plot(ax=ax)

# Construct the forecasts
fcst = pred.summary_frame()
fcst['mean'].plot(ax=ax, style='k--')
ax.fill_between(fcst.index, fcst['mean_ci_lower'], fcst['mean_ci_upper'], color='k', alpha=0.1);
```

```python
#ERROR CUADRADO MEDIO
#diferencia absoluta promedio entre la serie real y la predicha
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(serie, casos_predichos))
print('Test RMSE: %.3f' % rmse)
```

```python
#MAE: ERROR ABSOLUTO MEDIO
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(serie, casos_predichos)
print('MAE: %.3f' % mae)
```

```python
#ERROR EN PORCENTAJE DE LA MEDIA DE LOS ERRORES
def SMAPE (forecast, actual):
    """Returns the Symmetric Mean Absolute Percentage Error between two Series"""
    masked_arr = ~((forecast==0)&(actual==0))
    diff = abs(forecast[masked_arr] - actual[masked_arr])
    avg = (abs(forecast[masked_arr]) + abs(actual[masked_arr]))/2
    print('SMAPE Error Score: ' + str(round(sum(diff/avg)/len(forecast) * 100, 2)) + ' %')
    return sum(diff/avg)/len(forecast) * 100
```

```python
SMAPE(casos_predichos,serie)
```
